# Binomial Option Pricing

```python
import numpy as np
from math import exp, sqrt
import pandas as pd
from BinomialPricing import *
import matplotlib.pyplot as plt
```

```python
# Initialise parameters
S0 = 50      # initial stock price
K = 50       # strike price
T = 10       # time to maturity in years
r = 0.02     # annual risk-free rate
N = 10       # number of time steps
sigma = 0.2
u,d,q = up_down(sigma, N,T,r)
eu_option_pricing(K, T, S0, r, N, u,d,q,'call')
```

```python
steps=[i for i in range(1,1000)]

prices=[]
for N in steps:
    u,d,q = up_down(sigma, N,T,r)
    prices.append(eu_option_pricing(K, T, S0, r, N, u,d,q,'call'))
```

```python
# Plotting the graph
plt.plot(steps, prices, marker='o', linestyle='-')
plt.title('Call Option Price vs. Number of Steps')
plt.xlabel('Number of Steps')
plt.ylabel('Call Option Price')
plt.xscale('log')  # Set x-axis to a logarithmic scale for better visualization
plt.grid(True)
plt.show()
```

```python
steps=[1,10,50,100,250,500,1000,5000,10000]

prices=[]
for N in steps:
    u,d,q = up_down(sigma, N,T,r)
    prices.append(eu_option_pricing(K, T, S0, r, N, u,d,q,'put'))

for i,j in zip(range(len(prices)),steps):
    print(f"Put option price for n = {j}: {prices[i]:.4f}")
```
